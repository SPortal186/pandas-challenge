```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

```python
players_data = len(purchase_data["SN"].value_counts())
total_data = pd.DataFrame({"Total Players": [players_data]})
total_data
```

```python
item_count = len(purchase_data["Item ID"].unique())
mean_price = purchase_data["Price"].mean().round(2)
total_purchases = purchase_data["Purchase ID"].count()
total_money = purchase_data["Price"].sum()

summary_data = pd.DataFrame({"Number of Unique Items": [item_count],
                            "Average Price": "$"+str(mean_price),
                            "Number of Purchases": [total_purchases],
                            "Total Revenue": "$"+str(total_money)})

summary_data
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#To get values for SN
gender_data = purchase_data.groupby("SN")

# Need to create unique counts for Gender
gender_data_df = pd.DataFrame(gender_data["Gender"].unique().str[0].value_counts())

#Culculate percentages of players
gender_data_df["Percentage of Players"]=(gender_data_df["Gender"]/players_data*100).map("{:,.2f}%".format)

#Creating Total Counts
gender_data_df = gender_data_df.rename(columns={"Gender":"Total Count"})
                                                                    
gender_data_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Make groups of users
spenders_data = purchase_data.groupby("SN")

#Start calculations to obtain the basic results
#Purchases per ID
purchase_count = spenders_data["Purchase ID"].count() 
#Mean per spendar
ave_purchase_count = spenders_data["Price"].mean().map("${:,.2f}".format)
#Total per spender
total_purchase_count = spenders_data["Price"].sum()

#Create DataFrame with titles
summary_spenders_data = pd.DataFrame({"Purchase Count": purchase_count,
                            "Average Purchase Price": ave_purchase_count, 
                            "Total Purchase Value": total_purchase_count}) 
                          

#Descending order
summary_spenders_data_desc=summary_spenders_data.sort_values(['Total Purchase Value'], ascending=False)
#To format summary
summary_spenders_data_desc['Total Purchase Value'] = summary_spenders_data_desc['Total Purchase Value'].map('${:.2f}'.format)
summary_spenders_data_desc.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Item ID and Item Name
popular_item = purchase_data.groupby(["Item ID", "Item Name"])

#To obtain purchase count
purchase_popular_item = popular_item["Item ID"].count()
#Culculate average item price
mean_popular_price = popular_item["Price"].mean().map("${:,.2f}".format)
#Total purchase value
total_purchase_value = popular_item["Price"].sum()

#Create DataFrame with titles
summary_popular_item = pd.DataFrame({"Purchase Count": purchase_popular_item,
                            "Item Price": mean_popular_price, 
                            "Total Purchase Value": total_purchase_value}) 
#Descending order
summary_popular_item_desc=summary_popular_item.sort_values(['Purchase Count'], ascending=False)
#To format summary
summary_popular_item_desc['Total Purchase Value'] = summary_popular_item_desc['Total Purchase Value'].map('${:.2f}'.format)
summary_popular_item_desc.head()                         
```
